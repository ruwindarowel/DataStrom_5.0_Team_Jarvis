```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.base import TransformerMixin, BaseEstimator
from sklearn import set_config

set_config(transform_output = "pandas")

numeric_transfomer = StandardScaler()
nominal_transformer = OneHotEncoder(sparse_output=False)

preprocessor = ColumnTransformer(transformers=[
    ('num',numeric_transfomer,quants),
    ('nomi',nominal_transformer,nominal)
])


pipeline=Pipeline(steps=[
    ('preprocessor',preprocessor)
])

transformed_x_train = pipeline.fit_transform(x_train)

transformed_x_test = pipeline.fit_transform(x_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(multi_class='multinomial', solver='lbfgs')
log_reg_model=logreg.fit(transformed_x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_preds_log_reg_training = log_reg_model.predict(transformed_x_train)
y_preds_log_reg_testing = log_reg_model.predict(transformed_x_test)

accuracy_train_log_reg=accuracy_score(y_train, y_preds_log_reg_training, normalize=True)
accuracy_test_log_reg=accuracy_score(y_test, y_preds_log_reg_testing, normalize=True)

print(f"Training accuracy is, {accuracy_train_log_reg}")
print(f"Testing accuracy is, {accuracy_test_log_reg}")
```

## Using Feature Importance

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.base import TransformerMixin, BaseEstimator
from sklearn import set_config

set_config(transform_output = "pandas")


#Defining the Columns Dropper
#set_config(transform_output = "pandas") outputs the data as a dataframe instead of a numpy array
class DropColumnsTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, columns_to_drop):
        self.columns_to_drop = columns_to_drop
        
    def fit(self, X):
        return(self)
    
    def transform(self,X):
        return X.drop(self.columns_to_drop, axis=1)
    
#Defining the columns to encode
columns_to_encode = ['outlet_city']
columns_to_scale =  ['luxury_sales', 'fresh_sales', 'dry_sales']
columns_to_drop = ['onehot__outlet_city_kalmunai',
                   'onehot__outlet_city_colombo',
                   'onehot__outlet_city_galle',
                   'onehot__outlet_city_kandy',
                   'onehot__outlet_city_batticaloa',
                   'onehot__outlet_city_trincomalee',
                   'onehot__outlet_city_jaffna',
                   'onehot__outlet_city_gampaha',
                   'onehot__outlet_city_negombo',
                   'onehot__outlet_city_katunayake',
                   'onehot__outlet_city_dehiwala-mount lavinia',
                   'onehot__outlet_city_homagama',
                   'onehot__outlet_city_kaduwela',
                   'onehot__outlet_city_kelaniya',
                   'onehot__outlet_city_kotte',
                   'onehot__outlet_city_moratuwa',
                   'onehot__outlet_city_nuwara eliya',
                   'onehot__outlet_city_panadura',
                   'onehot__outlet_city_peliyagoda',
                   'onehot__outlet_city_wattala']

#Defining the preprocessor pipeline
pipeline = Pipeline([
    ('onehot_encoder', ColumnTransformer(
        transformers=[
            ('onehot',OneHotEncoder(sparse_output=False),columns_to_encode),
            ('scaler',StandardScaler(),columns_to_scale)
        ]
    )),
       ('column_dropper', DropColumnsTransformer(columns_to_drop))
])

transformed_data = pipeline.fit_transform(x_train)
transformed_data_test = pipeline.fit_transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

r_forrest = RandomForestClassifier(random_state=100)
r_forrest.fit(transformed_data,y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score

y_preds_r_forrest_best_training = r_forrest.predict(transformed_data)
y_preds_r_forrest_best_testing = r_forrest.predict(transformed_data_test)


#f1=f1_score(y_test,y_preds_r_forrest_best)
accuracy_train=round(accuracy_score(y_preds_r_forrest_best_training, y_train,normalize=True),7)
accuracy_test=round(accuracy_score(y_preds_r_forrest_best_testing, y_test,normalize=True),7)

print(F"Train accuracy is, {accuracy_train}")
print(f"Test accuracy is, {accuracy_test}")
#print(f"F1 score is, {f1}")
```

```python
from sklearn.metrics import confusion_matrix

cnf_matrix = confusion_matrix(y_test, y_preds_r_forrest_best_testing)
cnf_matrix

class_names=[1,2,3,4,5,6] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)

sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```
